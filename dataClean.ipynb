```python

#异常值的处理
for i in range(0,data.shape[0]):
    
    d = str(data.iloc[i].values[0])
    
    #print(d)
    
    #处理以>开头的数据
    if(d.startswith('>') or d.startswith('<') or d.startswith('≈') or d.startswith('~')):
        data.iloc[i] = d[1:]
        print(d)
    if(d.startswith('<=')):
        data.iloc[i] = d[2:]
        print(d)
    
    #处理以%结尾的数据
    if(d.endswith('%')):
        data.iloc[i] = float(d[:-1])/100
        print(d)
        
    #处理以°C结尾的数据
    if(d.endswith('°C') or d.endswith('**')):
        data.iloc[i] = float(d[:-2])
        print(d)
    #处理以*结尾的数据
    if(d.endswith('*')):
        data.iloc[i] = float(d[:-1])
        print(d)
        
    #处理以degC结尾的数据
    if(d.endswith('degC')):
        data.iloc[i] = d[:-4]
        print(d)
        
    #处理包含+—的数据    
    if(d.count('+-')):
        ds = d.split('+-')
        data.iloc[i] = ds[0]
        print(ds[0])
    if( d.count('±')):
        ds = d.split('±')
        data.iloc[i] = ds[0]
        print(ds[0])
        
    #处理以 [123] 的数据
    if(d.startswith('[')):
        data.iloc[i] = d[1:-1]
        print(d[1:-1])
        
    #其他的数据 A-B 或 A(B)或 A/B/C 的数据取平均值d
    ms = re.findall(r'([\d+\.]+)', d)
    m = 0.0
    for i in ms:
        m += float(i)
    avg = m/len(ms)
    #data.iloc[i] = avg
    print(avg)

    
    
    
    
    
    
   
    
    print(d) 
    
#将处理后的数据存储到原文件中
# writer = pd.ExcelWriter("E:\jupyter-notebook\learn\DataClean\data_sort.xlsx")
# data.to_excel(writer ,sheet_name=1)
# writer.save()
# writer.close()
```
